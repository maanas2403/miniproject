```python
from googletrans import Translator
import pandas as pd
```

```python
def translate_to_english(row):
    username = row['users']
    text = row['captions']  # Assuming 'text_column' is the name of your column containing text in multiple languages
    translator = Translator()
    try:
        translation = translator.translate(text, dest='en')
        print(f"Username: {username}, Translated Text: {translation.text}")
        return translation.text
    except Exception as e:
        print(f"Error translating text for username {username}: {e}")
        return None
```

```python
df=pd.read_csv('Top100.csv')
```

```python
df
```

```python
df['translated_text'] = df.apply(translate_to_english, axis=1)
```

```python
df.to_csv('traslated.csv')
```
