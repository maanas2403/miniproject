```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_topics(sentence):
    if pd.isna(sentence):
        return ['None']
    doc = nlp(sentence)
    nouns = [token.lemma_ for token in doc if token.is_alpha and not token.is_stop and token.pos_ == 'NOUN']
    return nouns
```

```python
df['topics'] = df['translated_text'].apply(extract_topics)
```

```python
df
```

```python
df = df[df['topics'].apply(lambda x: bool(x))]
```

```python
df
```

```python
from collections import Counter
```

```python
flat_topics = [topic for sublist in df['topics'] for topic in sublist]
```

```python
topic_counts = Counter(flat_topics)
```

```python
top_10_topics = dict(topic_counts.most_common(10))
top_10_topics
```

```python
plt.figure(figsize=(20, 10))
x = np.array([topic for topic, count in topic_counts.most_common(10)])
sns.barplot(x=top_10_topics.keys(),y= top_10_topics.values(),hue=x,data=top_10_topics,palette='magma')
plt.title('Top 10 Topics and Their Counts',fontdict={'size':20})
plt.xticks(fontsize=20,rotation=90)
plt.yticks(fontsize=20)
plt.xlabel('Topics',fontdict={'size':20})
plt.ylabel('Count',fontdict={'size':20})
plt.show()
```

```python
from collections import defaultdict
topic_likes = defaultdict(int)
unique_topics = set()
```

```python
for index, row in df.iterrows():
    unique_topics.update(row['topics'])
    for topic in unique_topics:
        topic_likes[topic] += row['Likes']
```
